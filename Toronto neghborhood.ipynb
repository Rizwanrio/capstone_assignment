```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate

res=requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup=BeautifulSoup(res.content,'lxml')
table=soup.find_all('table')[0]
df= pd.read_html(str(table))[0]

postalcode= df['Postal Code'].tolist()
borough=df["Borough"].tolist()
neighborhood=df["Neighborhood"].tolist()
```

```python
df_tuples=zip(postalcode,borough,neighborhood)
df_data=pd.DataFrame(df_tuples,columns=('Postal Code','Borough','Neighborhood'))
```

```python
df_drop=df_data[df_data.Borough!='Not assigned']
df_new=df_drop.sort_values(by=('Postal Code'),ascending=True)
df_new.reset_index(drop=True,inplace=True)
```

```python
df_ll=pd.read_csv('http://cocl.us/Geospatial_data')
df_ll.sort_values(by=('Postal Code'),ascending=True)
```

```python
df_new['Latitude']=df_ll['Latitude']
df_new['Longitude']=df_ll['Longitude']
df_new
```

```python
address = 'Toronto, tun'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Manhattan are {}, {}.'.format(latitude, longitude))
#latitude = 43.6532
#longitude =79.3832
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_new['Latitude'], df_new['Longitude'], df_new['Borough'], df_new['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
